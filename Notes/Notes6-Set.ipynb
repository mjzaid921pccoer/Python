# Data Type : set {}
* Unordered Collection of Unique Items
* Defined by {} braackets separated by commas
* Mathematical Set operations like Union, Intersection, etc can be performed
* No Duplicates Items
* No Slicing for accessing, Hence **unindexed collection**
* immutable i.e current set elements cannot be altered, Note that some method can add/update new elements in existing set. 

````python
# Set of players playing Cricket
Cricket = { "Abhi", "Bhaskar", "Chetan", "David", "Gitesh", "Manav", "Rajeev"}

# Set of players playing Football
Football = { "Bhaskar", "Manav", "Mayuresh","Rajeev", "Siddhesh", "Omkar", "Piyush"}
````

````python
for player in Cricket:
    print(player)
````

````python
for player in Football:
    print(player)
````

### ```set.add(Item)``` Method
* adds one element to set

````python
Cricket.add("Rahul")
Football.add("Rohan")

print("Cricket",Cricket)
print("Football",Football)
````

### ```set.update(Iterable)``` Method
* adds all iterable elements to set 

Iterable maybe : set, string, list, tuple, etc 

````python
cricket_junior = {"nikhil", "atharva", "ravi"}
football_junior = {"ritwik", "sahil"}

Cricket.update(cricket_junior)

Football.update(football_junior)

print("Cricket",Cricket)
print("Football",Football)
````

### ```set.discard(Item)``` Method
* remove the Item if exist in set 

````python
print("cricket_junior",cricket_junior)

cricket_junior.discard("ravi")

print("cricket_junior",cricket_junior)
````

### ```set.remove(Item)``` Method
* remove the Item if exist in set else raise error

````python
print("football_junior",football_junior)

football_junior.remove("ritwik")

print("football_junior",football_junior)
````

### ```set.pop()``` Method
* remove the last Item from set : But Set are Unordered Collection hence we actually never predict which Item would be popped hence not prefered

### ```set.clear()``` Method
* remove all the Item from set object

````python
print("cricket_junior",cricket_junior)

cricket_junior.clear()

print("cricket_junior",cricket_junior)
````

## SET Operations

````python
#!pip install matplotlib_venn
from matplotlib_venn import venn2_unweighted,venn2,venn2_circles,venn3
from matplotlib import pyplot as plt
%matplotlib inline

def draw_vennDiagram(sets=((1,2,3,4,5,6),(4,5,6,7,8,9)),set_labels=('set A','set B') ,set_colours=('orange','blue'),operation=0):
    set1, set2 = sets
    label1, label2 = set_labels
    colour1, colour2 = set_colours
    s1ns2 = len(set1.intersection(set2))
    s1 = len(set1)
    s2 = len(set2)

    v=venn2( [set1, set2],
           set_labels = (label1,label2),
           set_colors = (colour1,colour2), alpha = 0.7
         )
    venn2_circles(subsets = (s1-s1ns2, s2-s1ns2, s1ns2)); # Black Outline
    #plt.title('Players')
    
    if(operation=='union'):
        v.get_patch_by_id('10').set_color('green')
        v.get_patch_by_id('11').set_color('green')
        v.get_patch_by_id('01').set_color('green')
    elif(operation=='intersection'):
        v.get_patch_by_id('10').set_color('white')
        v.get_patch_by_id('11').set_color('green')
        v.get_patch_by_id('01').set_color('white')
    elif(operation=='difference1'):
        v.get_patch_by_id('10').set_color('green')
        v.get_patch_by_id('11').set_color('white')
        v.get_patch_by_id('01').set_color('white')
    elif(operation=='difference2'):
        v.get_patch_by_id('10').set_color('white')
        v.get_patch_by_id('11').set_color('white')
        v.get_patch_by_id('01').set_color('green')
    elif(operation=='symmetric-difference'):
        v.get_patch_by_id('10').set_color('green')
        v.get_patch_by_id('11').set_color('white')
        v.get_patch_by_id('01').set_color('green')
        
    
    
    
    
    plt.show()

draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'))
````

# **```set1.union(set2)```**

* returns SET Union operation combines both set, 

```.update()``` gives similar result on any iterables while union is for set.


````python
draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'),operation='union')
````

````python
print("# Students who play Game")

Players = Cricket.union(Football)
print(Players)
````

# **```set1.intersection(set2)```**

* returns SET Intersection operation : Common Items from both set.


````python
draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'),operation='intersection')
````

````python
print("# Students who play both Cricket & Football Game")

Players = Cricket.intersection(Football)
print(Players)
````

# **```set1.intersection_update(set2)```**

* updates set1 with result after SET Intersection operation : Common Items from both set.

````python
print("# Set A as : Students who play both Cricket & Football Game")

Original_Cricket = Cricket.copy()
print(Cricket)

Cricket.intersection_update(Football)

print(Cricket)
````

````python
# restoring values

Cricket = Original_Cricket.copy()

# releasing memory for variable
del Original_Cricket

print(Cricket)
````

# **```set1.difference(set2)```**

* returns set1 items discarding intersection & set2 items.


````python
draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'),operation='difference1')
````

````python
print("# Students who play only Cricket")

Players = Cricket.difference(Football)
print(Players)
````

````python
draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'),operation='difference2')
````

````python
print("# Students who play only Football")

Players = Football.difference(Cricket)
print(Players)
````

# **```set1.difference_update(set2)```**

* updates set1 with result after SET difference operation : set1 items discarding intersection & set2 items..

````python
print("# Set A as : Students who play only Cricket Game")

Original_Cricket = Cricket.copy()
print(Cricket)

Cricket.difference_update(Football)

print(Cricket)
````

````python
# restoring values

Cricket = Original_Cricket.copy()

# releasing memory for variable
del Original_Cricket

print(Cricket)
````

# **```set1.symmetric_difference(set2)```**

* returns **complement of SET Intersection operation** : Items in set1 and set2 but not Common Items from both set.


````python
draw_vennDiagram(sets=(Cricket,Football),set_labels=('Cricket','Football'),operation='symmetric-difference')
````

````python
print("# Students who play Single Game Cricket or Football")

Players = Cricket.symmetric_difference(Football)
print(Players)
````

# **```set1.symmetric_difference_update(set2)```**

* set1 is updated to **complement of SET Intersection operation** : Items in set1 and set2 but not Common Items from both set.


# ```set1.isdisjoint(set2)```

* returns True if both set do not have anything common

````python
alice_hobby = {"Singing", "Trekking", "Reading"}
bob_hobby = {"Cricket", "Movies", "Vlogging"}

draw_vennDiagram(sets=(alice_hobby,bob_hobby),set_labels=('Alice','Bob'))

print("#Do alice & bob hobbies are apart from each other?")
print(alice_hobby.isdisjoint(bob_hobby))
````

# ```set1.issubset(set2)```
* returns True if set1 is subset of set2

````python
charlie_hobby = {"Singing","Reading"}

draw_vennDiagram(sets=(alice_hobby,charlie_hobby),set_labels=('Alice','Charlie'))

print(clarlie_hobby.issubset(alice_hobby))
````

# ```set1.issuperset(set2)```
* returns True if set1 is superset of set2


````python
draw_vennDiagram(sets=(alice_hobby,charlie_hobby),set_labels=('Alice','Charlie'))

print(alice_hobby.issuperset(clarlie_hobby))
print(clarlie_hobby.issuperset(alice_hobby))
````
